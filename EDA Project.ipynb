# Outlier Detection & Removal

An outlier is a point or set of points that are different from other points. Sometimes they can be very high or very low. It's often a good idea to detect and remove the outliers. Because outliers are one of the primary reasons for resulting in a less accurate model. Hence it's a good idea to remove them. The outlier detection and removing that I am going to perform is called IQR score technique. Often outliers can be seen with visualizations using a box plot. Shown below are the box plot of MSRP, Cylinders, Horsepower and EngineSize. Herein all the plots, you can find some points are outside the box they are none other than outliers. The technique of finding and removing outlier that I am performing in this assignment is taken help of a tutorial from[ towards data science](https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba).

## Column-by-Column IQR

**2016**

```python
q1 = np.percentile(df1['Cases Reported'], 25)
# Calculate the third quartile (75th percentile)
q3 = np.percentile(df1['Cases Reported'], 75)
# Calculate the interquartile range (IQR)
iqr = q3 - q1
print("Interquartile Range for Cases Reported:", iqr)

# This is a method to find IQR for individual columns
```

```python
a1=q1-1.5*iqr
a2=q3+1.5*iqr
print("Normal data range for Cases Reported:",a1,",",a2)
```

```python
# create a dataframe with average cost by city
df_cost_city = df1.groupby('State/UT')['Cases Reported'].mean().reset_index()

# sort the dataframe by cost
df_cost_city = df_cost_city.sort_values('Cases Reported')

# plot the data as a horizontal bar chart
plt.figure(figsize=(12, 8))
plt.barh(df_cost_city['State/UT'], df_cost_city['Cases Reported'], color='#FFC107')
plt.title('Comparison of cases reported in State/UT in 2016', fontsize=18)
plt.xlabel('Cases Reported', fontsize=14)
plt.ylabel('State/UT', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Piecharts

```python
#Group the data by Category (State/UT) and sum the Cases Reported
state_ut_cases = df3.groupby('Category')['Cases Reported'].sum().reset_index()

# Create a pie chart
plt.pie(state_ut_cases['Cases Reported'], labels=state_ut_cases['Category'], autopct='%1.1f%%')
plt.title('Rape Cases in States vs Union Territories')
plt.show()
```

```python
# create a dataframe with average victim count by state
df_vic = df1.groupby('State/UT')['Total Victims'].mean().reset_index()

# sort the dataframe percentage of rape victims
df_vic = df_vic.sort_values('Total Victims')

# create a pie chart
plt.figure(figsize=(8,8))
plt.pie(df_vic['Total Victims'], labels=df_vic['State/UT'], autopct='%1.1f%%', startangle=90, explode=[0.1]*len(df_vic))
plt.title('Comparison of Rape Victims Across States in 2016')
plt.show()
```

```python
# Calculate the total number of women victims of rape (above 18 years)
total_women_victims = df2['Women Victims of Rape (Above 18 Yrs) - Total Women/Adult Victims'].sum()

# Calculate the number of women victims of rape for each age group
women_victims_18_29 = df2['Women Victims of Rape (Above 18 Yrs) - 18 Years & Above - Below 30 Years'].sum()
women_victims_30_44 = df2['Women Victims of Rape (Above 18 Yrs) - 30 Years & Above - Below 45 Years'].sum()
women_victims_45_59 = df2['Women Victims of Rape (Above 18 Yrs) - 45 Years & Above - Below 60 Years'].sum()
women_victims_60_plus = df2['Women Victims of Rape (Above 18 Yrs) - 60 Years & Above'].sum()

# Create a list of labels and sizes for the pie chart
labels = ['18-29 years', '30-44 years', '45-59 years', '60 years and above']
sizes = [women_victims_18_29, women_victims_30_44, women_victims_45_59, women_victims_60_plus]

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Women Victims of Rape (Above 18 years) by Age Group in 2017')
plt.show()
```
